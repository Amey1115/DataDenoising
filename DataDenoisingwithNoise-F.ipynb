```python
import numpy as np
import matplotlib.pyplot as plt
plt.rcParams['figure.figsize'] = [16,12]
dt=0.001
t=np.arange(0,1,dt)
f=np.sin(2*np.pi*50*t)+ np.sin(2*np.pi*120*t)+1.5*np.random.randn(len(t))
plt.plot(t,f,color='c',linewidth=1.5,label='Signal')

plt.xlim(t[0],t[-1])
```

```python
n=len(t)
fhat=np.fft.fft(f,n)
psd=fhat * np.conj(fhat) /n
freq=(1/(dt*n))*np.arange(n)
l=np.arange(1,np.floor(n/2),dtype='int')


plt.plot(freq[l],np.real(psd[l]),color='c',linewidth=2,label='Signal')
plt.xlim(freq[l[0]],freq[l[-1]])
plt.legend()
plt.show()
```
